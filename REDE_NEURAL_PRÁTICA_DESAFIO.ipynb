# VOCÊ RECEBERÁ UMA BASE REFERENTE A CHURN, ONDE A SUA VARIÁVEL TARGET SERÁ A CHURN. FAÇA O TREINAMENTO E USE A ATIVAÇÃO LOGISTIC (SIGMOIDE), E ANALISAR A ACURÁCIA DA SAÍDA:

```python
import pandas as pd
```

```python
df = pd.read_excel("/content/churn_data.xlsx")

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# NESSE CASO A COLUNA customerID NÃO É NECESSÁRIA, PODEMOS EXCLUI-LA:

df.drop(columns=["customerID"])
```

```python
# EXCLUIR REGISTROS NULOS DA BASE DE DADOS:

df = df.dropna()
```

```python
df.isnull().any()
```

```python
# TRANSFORMAR VARIÁVEIS OBJECT EM NUMÉRICAS:

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors = "coerce")
```

```python
# SEPARAR VARIÁVEIS NUMÉRICAS E CATEGÓRICAS:

cat_vars = ["PhoneService", "Contract", "PaperlessBilling","PaymentMethod"]

num_vars = ["tenure", "MonthlyCharges", "TotalCharges"]

modeling_vars = cat_vars + num_vars
```

```python
# DISTRIBUIR X E Y:

x = df.filter(modeling_vars).copy()

y = df["Churn"].copy()

y = pd.DataFrame(y)
```

```python
# TRANSFORMAR COLUNA "CHURN" EM FORMATO BINÁRIO:

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(y.Churn)

y.Churn = le.transform(y.Churn)
```

```python
y
```

```python
# TRANSFORMAR VARIÁVEIS CATEGÓRICAS DO X EM DUMMIES:

x = pd.get_dummies(x)
```

```python
x
```

```python
# VERIFICAR A DISTRIBUIÇÃO DA NOSSA VARIÁVEL TARGET:

df["Churn"].value_counts() # resultado: não 5173, sim 1869
```

```python
# SEPARAR TREINO E TESTE:

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test, = train_test_split(x, y, test_size=0.2)
```

```python
# FAZER NORMALIZAÇÃO DAS NOSSAS VARIÁVEIS:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
# TRANSFORMAR TUDO EM DATAFRAME:

X_train = pd.DataFrame(X_train_scaled, columns = X_train.columns)

X_test = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
# IMPLEMENTAR A REDE NEURAL:

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

clf = MLPClassifier(hidden_layer_sizes = (100, 50, 20), activation = "logistic")
# hiperparâmetro (camadas escondidas = "hidden_layer_sizes = (100, 50, 20)") e a função de ativação (sigmoide = "activation = "logistic")

clf.fit(X_train, y_train)
```

```python
# VERIFICAR A ACURÁCIA E O F1 DO MODELO:

y_pred = clf.predict(X_test)
print(f"Acurácia: {accuracy_score(y_test, y_pred)*100}%")
print(f"F1 Score: {f1_score(y_test, y_pred)*100}%")
```
